# Using Beautiful Soup to scrape text off of a website


```python
import requests
from bs4 import BeautifulSoup
```

## Make a request to the website and get the html

```python
r = requests.get("https://en.wikipedia.org/wiki/Harvey_Jacobs")
r.content
```

## Use the 'html.parser' to parse the page

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

## Find the paragraphs with 'find_all' to find paragraph tags

```python
paragraphs = soup.find_all('p')
```

```python
for paragraph in paragraphs:
    print(paragraph.text)
```

## Define a function that looks for text in the html that we just scraped

```python
def search_for_text_in_html(desired_text):
    if desired_text in soup.text.strip():
        print("The word or words '" + desired_text + "' is in the html")
    else:
        print("The word or words '" + desired_text + "' is not in the html")

```

## Use the function to search for text

```python
search_for_text_in_html("Beautiful Soup")
```
